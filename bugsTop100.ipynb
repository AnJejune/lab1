```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
```

```python
print(dt.now())
```

```python
targetSite = 'https://music.bugs.co.kr/chart'
#requests 모듈의 get() 메소드로 타켓 사이트의 정보를 얻어온다(요청한다)
request = requests.get(targetSite)
#print(request)
html = request.text
#print(html)
soup = BeautifulSoup(html, 'html.parser') #객체로 변환 soup는 객체
```

```python
#노래 제목 크롤링 
#노래 제목은 'class' 속성이 'title'인 <p> 태그 내부에 있다.
titles = soup.findAll('p', {'class':'title'})
#print(len(titles))
for title in titles:
    #print(title.text.strip())
    #print(title.text.split('\n'))
    #print(title.text.split('\n')[1])
    print(title.text.strip().split('\n')[0])
```

```python
#아티스트 크롤링
#가수 이름은 
artists = soup.findAll('p',{'class':'artist'})
for artist in artists:
    print(artist.text.strip().split('\n')[0])
```

```python
print(dt.now())
for i in range(len(titles)):
    title = titles[i].text.strip().split('\n')[0]
    artist = artists[i].text.strip().split('\n')[0]
    print('{0:3d}등 {1} - {2}'.format(i + 1, artist, title))
```

```python
# 크롤링한 결과를 텍스트 파일에 저장한다.
file = open('./bugsTop100.txt','w',-1,'UTF-8') #쓰기 목록 해당 파일이 없으면 새로 만들고 있으면 덮어쓰기

file.write('{} 현재 Bugs 뮤직 실시간 TOP 100\n'.format(dt.now()))
for i in range(len(titles)):
    title = titles[i].text.strip().split('\n')[0]
    artist = artists[i].text.strip().split('\n')[0]
    file.write('{0:3d}등 {1} - {2} \n'.format(i + 1, artist, title))

#파일을 열었으면 반드시 닫아라
file.close()
print("파일종료")
```

```python
#텍스트 파일에 저장된 데이터를 읽어서 출력한다.
try:
    file = open('./bugsTop100.txt','r',-1,'UTF-8')
    # readlines() 메소드는 텍스트 파일의 내용 전체를 한꺼번에 읽어서 리스트로 얻어온다.
    lines = file.readlines()
    for line in lines:
        print(line.strip())
    
except FileNotFoundError:
    print("해당 파일을 디스크에서 찾을 수 없습니다.")
```
